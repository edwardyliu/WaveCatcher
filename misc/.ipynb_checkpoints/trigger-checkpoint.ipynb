```python
import numpy as np
import pandas as pd
from scipy.signal import butter, lfilter

def butter_bandpass(lowcut, highcut, fs, order=5):
    nyq = 0.5 * fs
    low = lowcut / nyq
    high = highcut / nyq
    b, a = butter(order, [low, high], btype='band')
    return b, a


def butter_bandpass_filter(data, lowcut, highcut, fs, order=5):
    b, a = butter_bandpass(lowcut, highcut, fs, order=order)
    y = lfilter(b, a, data)
    return y
```

```python
rawdf = pd.read_csv('1552794169380_recording_MusicA.csv')
TP9 = rawdf['TP9']
TP10 = rawdf['TP10']
AF7 = rawdf['AF7']
AF8 = rawdf['AF8']
```

```python
# Load the Dataset
fs = 128
data = TP9.values
print(data.shape)
print(data)
```

```python
# Clean the Dataset
data = data[~np.isnan(data)]
dataFiltered = butter_bandpass_filter(data, 1e-5, 45, fs, order=6)
```

```python
fft_vals = np.absolute(np.fft.rfft(data))
fft_freq = np.fft.rfftfreq(len(data), 1.0/fs)
eeg_bands = {
    'Delta': (0, 4),
    'Theta': (4, 8),
    'Alpha': (8, 12),
    'Beta' : (12, 30),
    'Gamma': (30, 45)
}
```

```python
eeg_band_fft = dict()
for band in eeg_bands:
    freq_itv = np.where((fft_freq >= eeg_bands[band][0]) & (fft_freq <= eeg_bands[band][1]))[0]
    eeg_band_fft[band] = np.mean(fft_vals[freq_itv])
```

```python
df = pd.DataFrame(columns=['band', 'val'])
df['band'] = eeg_bands.keys()
df['val'] = [eeg_band_fft[band] for band in eeg_bands]
ax = df.plot.bar(x='band', y='val', legend=False)
ax.set_xlabel("EEG band")
ax.set_ylabel("Mean band Amplitude")
df.head(10)
theta = df['val'][1]
alpha = df['val'][2]
beta = df['val'][3]
```

```python
heur1 = (theta+alpha)/beta
heur2 = theta/alpha
heur3 = alpha
print(heur1)
print(heur2)
print(heur3)
```
